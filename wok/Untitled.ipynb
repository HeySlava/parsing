```python
import pandas as pd
import json
```

```python
with open('./wok_tula.json') as f:
    data = json.load(f)

df_main = pd.DataFrame()
for key in data['wok_tula']['section'].keys():
    df = pd.DataFrame(data['wok_tula']['section'][key]).T.reset_index().rename(columns={'index': 'name'})
    df_main = df_main.append(df, ignore_index=True)
df_main.head(2)
```
